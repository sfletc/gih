## Align

````python
reads = ""
for i in glob.glob('reads/processed/*.gz'):
    reads = reads + i + ","
reads = reads[:-1]
print(reads)
````

````python
! software/scram2 profile \
    -r ref/tri6.fa \
    -1 {reads} \
    -l 20,21,22,23,24 \
    -o test_align
````
